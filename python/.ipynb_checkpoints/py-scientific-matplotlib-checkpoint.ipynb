```python
import numpy as np
import matplotlib.pyplot as plt
```

MatLab 数学绘图工具

```python
%matplotlib inline
```

```python
x = np.arange(0, 6, 0.1)
y = np.sin(x)
print(y)
```

```python
plt.plot(x, y)
plt.show()
```

```python
x = np.arange(0, 6, .1)
y1 = np.sin(x)
y2 = np.cos(x)

plt.plot(x, y1, label='sin')
plt.plot(x, y2, label='cos', linestyle='--')
plt.xlabel('x')
plt.ylabel('y')
plt.title('sin & cos')
plt.legend()
plt.show()
```

```python
x = np.arange(0, 3 * np.pi, 0.1)
y1 = np.sin(x)
y2 = np.cos(x)

plt.subplot(2, 1, 1)
plt.plot(x, y1)
plt.title('sine')

plt.subplot(2, 1, 2)
plt.plot(x, y2)
plt.title('cosine')

plt.show()
```
